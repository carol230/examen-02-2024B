```python
import numpy as np
from scipy.optimize import fsolve
from scipy.optimize import bisect, newton

def polinomio(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

def polinomio_derivada(x):
    return 5*x**4 - 24*x**3 + 6*x**2 + 40*x - 27

raices_newton = []
for a in np.linspace(-10, 10, 10):
    try:
        root = newton(polinomio, a, fprime=polinomio_derivada)
        if not any(np.isclose(root, r, atol=1e-6) for r in raices_newton):
            raices_newton.append(root)
    except RuntimeError:
        pass
print(f"Soluciones: {raices_newton}")

# Comprobar las raíces 
verification1 = [polinomio(-2)]
print (verification1)
verification2 = [polinomio(1)]
print(f"Prueba de 1 0 valor real: {verification2[0]:.100f}")
verification3 = [polinomio(0.9999994521465998)]
print(f"Prueba: {verification3[0]:.100f}")
verification4 = [polinomio(5)]
print (verification4)
```
